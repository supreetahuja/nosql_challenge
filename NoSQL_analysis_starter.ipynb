### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

# Print the number of documents in the result

# Print the first 10 results
```

```python
# Define the aggregation pipeline
pipeline = [
    {
        "$match": {"scores.Hygiene": 0}
    },
    {       "$group": {
            "_id": "$LocalAuthorityName",  
            "count": {"$sum": 1} }
    },
    {
        "$sort": {
            "count": -1  
        }
    },
    {
        "$limit": 10}
]

# Execute the aggregation pipeline
result = establishments.aggregate(pipeline)

# Iterate through the results and print the count for each Local Authority
for doc in result:
    print("Local Authority:", doc["_id"])
    print("Number of establishments with hygiene score 0:", doc["count"])
    print("\n")
```

```python
# Execute the aggregation pipeline
result = establishments.aggregate(pipeline)

# Convert the aggregation result to a list to count the documents
result_list = list(result)

# Count the number of documents in the result
total_documents = len(result_list)

# Print the total number of documents
print("Total number of documents in the result:", total_documents)
```

```python
# Convert the result to a Pandas DataFrame

# Display the number of rows in the DataFrame

# Display the first 10 rows of the DataFrame
```
